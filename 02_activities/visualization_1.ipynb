### Seaborn Themes

The seaborn python library is well known for its grey background and its general styling. However, there are few other built in styles available: darkgrid, white grid, dark, white and ticks.

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
df = pd.DataFrame(data)
df.head()
```

```python
sns.boxplot(data=data)
plt.show()
```

```python
# White grid
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.title("whitegrid")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(data=data);
plt.title("darkgrid")
plt.show()
```

```python
sns.set_style("white")
sns.boxplot(data=data);
plt.title("white")
plt.show()
```

```python
sns.set_style("dark")
sns.boxplot(data=data);
plt.title("dark")
plt.show()
```

```python
# Ticks
sns.set_style("ticks")
sns.boxplot(data=data);
plt.title("new title")
plt.show()
```
