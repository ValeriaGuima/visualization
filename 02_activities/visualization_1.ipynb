```python
# Creating various plots from sample data
np.random.seed(613)
x = np.arange(50)
y=np.random.randint(0,100,50)

# making a scatterplot
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
# modifying the chart to a bar graph
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
# creating a lineplot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
#making a basic histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

### Plot Labels

```python
# Adding labels to the lat plot created
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
#defining font dictionaries
font1 ={'family' :'sans-serif', 'color':'blue','size':20}
font2 ={'family' :'monospace', 'color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time',fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start',fontdict = font2)
```

```python
# aligning labels in the chart 
font1 ={'family' :'sans-serif', 'color':'blue','size':20}
font2 ={'family' :'monospace', 'color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time',fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2, loc="center")
ax.set_xlabel('Years since start',fontdict = font2)
```

```python
#modifying data points
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,
           y,
           marker='*', #* or any other symbol
           color ='indigo')
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot (x,y,
        marker='*', #* or any other symbol
        color ='indigo',
        linestyle = '--',
        linewidth = '2')
```

```python
# modifying data points
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color ='#7425b9',
        linestyle = '--',
        linewidth = '2',
        markeredgecolor = '#fa9359',
        markerfacecolor='#000000')

```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color ='#7425b9',
        linestyle = '--',
        linewidth = '2',
        markeredgecolor = '#fa9359',
        markerfacecolor='#000000')

#adding grid lines to the axix object using.grid()
ax.grid(color = 'blue', linewidth =1, linestyle = 'dotted')
```

### Seaborn Themes

The seaborn python library is well known for its grey background and its general styling. However, there are few other built in styles available: darkgrid, white grid, dark, white and ticks.

```python
import seaborn as sns
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
df = pd.DataFrame(data)
df.head()
```

```python
sns.boxplot(data=data)
plt.show()
```

```python
# White grid
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.title("whitegrid")
plt.show()
```

```python
#adding a dark grid
sns.set_style("darkgrid")
sns.boxplot(data=data);
plt.title("darkgrid")
plt.show()
```

```python
sns.set_style("dark")
sns.boxplot(data=data);
plt.title("dark")
plt.show()
```
