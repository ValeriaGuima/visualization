```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'C:\Users\user\Projects\visualization\02_activities\assignments\earlyon_toronto_clean.csv')

# Count the number of programs per ward
programs_per_ward = data.groupby('ward_name').size().reset_index(name='program_count')

# Sort wards by the number of programs and select the top 10
top_wards = programs_per_ward.sort_values(by='program_count', ascending=False).head(10)

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Bar plot for the number of programs per ward
sns.barplot(x='program_count', y='ward_name', data=top_wards, palette='viridis')

# Add labels and title
plt.xlabel('Number of Programs')
plt.ylabel('Ward Name')
plt.title('Top 10 Wards by Number of EarlyON Programs')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Select only the relevant columns for the pairplot
pairplot_data = data[['ward_name', 'agency', 'lat', 'lng']]

# Plot a pairplot with hue as 'ward_name' to show different wards
sns.pairplot(pairplot_data, hue='ward_name', diag_kind='kde', height=2.5)

# Show the plot
plt.show()
```
